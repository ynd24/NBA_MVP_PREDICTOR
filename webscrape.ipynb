```python
import requests
import os
import shutil
```

```python
years = list(range(1991, 2023))
```

```python
url_start = "https://www.basketball-reference.com/awards/awards_{}.html"

for year in years:
    url = url_start.format(year) # replaces the curly bracks w year
    
    data = requests.get(url) # request the html data
    
    with open("mvp/{}.html".format(year), "w+") as f: # create new file with year in name, w+ as overwrite if necessary
        f.write(data.text) # write html to file
        
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
dfs = []

for year in years:
    with open("mvp/{}.html".format(year)) as f: # go through html files
        page = f.read() # store them in page
    soup = BeautifulSoup(page, 'html.parser') # parse html
    soup.find('tr', class_="over_header").decompose() # remove over header
    mvp_table = soup.find(id="mvp") # find table
    mvp = pd.read_html(str(mvp_table))[0] # convert to pd df
    mvp["year"] = year # add year column to keep track

    dfs.append(mvp)
    
```

```python
mvps = pd.concat(dfs) 
```

```python
mvps.to_csv("mvps.csv")
```

```python
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

```python
y2 = [2021, 2022] # adding 2 missing years
for y in y2:
    url = team_stats_url.format(y)
    
    data = requests.get(url)
    
    with open("team/{}.html".format(y), "w+") as f:
        f.write(data.text)
```

```python
for year in years:
    url = team_stats_url.format(year)
    
    data = requests.get(url) # get the team stats
    
    with open("team/{}.html".format(year), "w+") as f:
        f.write(data.text)
```

```python
dfs = []
for year in years:
    with open("team/{}.html".format(year)) as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="thead").decompose()
    e_table = soup.find_all(id="divs_standings_E")[0]
    e_df = pd.read_html(str(e_table))[0]
    e_df["Year"] = year
    e_df["Team"] = e_df["Eastern Conference"]
    del e_df["Eastern Conference"]
    dfs.append(e_df)
    
    w_table = soup.find_all(id="divs_standings_W")[0]
    w_df = pd.read_html(str(w_table))[0]
    w_df["Year"] = year
    w_df["Team"] = w_df["Western Conference"]
    del w_df["Western Conference"]
    dfs.append(w_df)
```

```python
teams = pd.concat(dfs)
```

```python
teams.to_csv("teams.csv")
```
