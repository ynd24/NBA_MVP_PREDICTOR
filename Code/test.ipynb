```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players["Player"].head(50)
```

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players["Player"].head(50)
```

```python
players.groupby(["Player", "year"]).get_group(("Greg Anderson", 1991))
```

```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "year"]).apply(single_team)
```

```python
players
```

```python
players.index = players.index.droplevel()
```
